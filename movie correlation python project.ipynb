```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=12,8
```

```python
#read in the data
df=pd.read_csv(r'C:\Users\annek\Downloads\movies.csv')
```

```python
#read in the data
df.head()
```

```python
#check for missing data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
#data types for our columns
df.dtypes
```

```python
#Handling missing data
df.isnull().sum()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sns.histplot([df.budget])
median=df['budget'].median()
df['budget'].fillna(median,inplace=True)
```

```python
df['rating'].fillna('no rating',inplace=True)
median=df['score'].median()
df['score'].fillna(median,inplace=True)
df['votes'].fillna(median,inplace=True)
df['gross'].fillna(median,inplace=True)
df['runtime'].fillna(median,inplace=True)
df['released'].fillna('none',inplace=True)
df['writer'].fillna('none',inplace=True)
df['star'].fillna('none',inplace=True)
df['country'].fillna('none',inplace=True)
df['company'].fillna('none',inplace=True)
```

```python
#change data types of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')
plt.show()
```

```python
#let's start looking at correlation
df.corr(method='pearson')#only works on numerical fields
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('budget vs gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')

plt.show()
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python

correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('budget vs gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
